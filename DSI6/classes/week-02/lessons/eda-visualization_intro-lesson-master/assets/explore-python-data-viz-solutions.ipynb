### 2. Research and implement a line and bar plot using `pandas`' `.plot()` functionality.

Consider just the first 10 rows of the data using the variables "volume_sold" and "2015_q1_sales."

Which plot type appears more useful for working with these variables?

```python
sales_10 = sales.iloc[0:10, :]
```

```python
sales_10.plot(x='volume_sold', y='2015_q1_sales', kind='line')
```

```python
sales_10.plot(x='volume_sold', y='2015_q1_sales', kind='bar')
```

**Answer:** _Which plot type appears more useful for working with these variables?_
>While the bar chart is probably more useful for visualizing the difference between our `2015_q1_sales` figures, neither is the best way to gain insight when comparing the two variables visually.
Remember, bar charts are most effective when comparing numerical data with categorical data. In this situation, both variables are numeric.
The line plot doesn't help us understand the x and y relationships in each row, as it is difficult to identify individual data points. Using a line in a data visualization is most effective when one of the variable's order matters, such as time.

### 4. Implement a `seaborn` "pairplot."

What do each of these elements represent? Is this more or less useful than the previous plot?

**Answer:** _What do each of these elements represent? Is this more or less useful than the previous plot?_

This is very useful for quickly discovering which variables have some kind of correlation during our exploratory data analysis. However, when you're looking solely at the `2015_margin` feature, the pairplot is more difficult to read and interpret than the single histogram.
