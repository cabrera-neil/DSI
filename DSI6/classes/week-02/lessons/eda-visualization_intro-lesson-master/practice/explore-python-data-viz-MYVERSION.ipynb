### Using a DataFrame and Matplotlib commands we can get fancy

```python

# set the title
df[['col1', 'col2']].plot(title='Col1 and Col2')

# move the legend
plt.legend(loc=4)

# x-axis labels
plt.xlabel('this is the x axis')

# y-axis labels
plt.ylabel('this is the y axis')
plt.yticks(np.arange(-3,3,0.5), 
           ['ciao {}'.format(i) for i in np.arange(-3,3,0.5)])
plt.show()
```

### Horizontal bar plot

```python
# A:

df['col1'].plot(kind='barh')
```

### Create a bar chart using col1 and col2.

- Do the same thing but with horizontal bars.

### We can use stacked bars
