### 1) Shape - smooth vs. "step"  
### 2) Position - centered on -100, 0, 10, 50, etc.  
### 3) Orientation - s-shape vs. z-shape  

# Load the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline 
plt.style.use('fivethirtyeight')
pd.set_option('display.max_columns', None)
```

# Define the logistic function

```python
def logistic(x, beta0=0, beta1=1):
    return 1 / (1 + np.e**-(beta0 + beta1 * x))
    #return 1 / (1 + np.e**-x)
```

b0 + b1\*x1 --> what is it similar to? 

# Plot a basic example

```python
x = np.linspace(-10,10,100)
y = [logistic(xi) for xi in x]
```

```python
plt.scatter(x,y)
```

# Let's change beta1

```python
for beta1 in [0.5,1,2,5,10]:  
    y = [logistic(xi, beta1=beta1) for xi in x]
    plt.plot(x,y)
    plt.title('beta1 = {}'.format(beta1))
    plt.show()
```

# Let's change beta0

```python
for beta0 in [-5,-1,0.1,1,5]:  
    y = [logistic(xi, beta0=beta0) for xi in x]
    plt.plot(x,y)
    plt.title('beta0 = {}'.format(beta0))
    plt.axvline(0, c='r')
    plt.show()
```

# What if my 0s are on the right and my 1s on the left?
### Use a negative beta1!

```python
BETA1 = 2
y = [logistic(xi, beta1=BETA1) for xi in x]
plt.plot(x,y)
plt.title('beta1 = {}'.format(BETA1))
plt.show()
```

```python
BETA1 = -2
y = [logistic(xi, beta1=BETA1) for xi in x]
plt.plot(x,y)
plt.title('beta1 = {}'.format(BETA1))
plt.show()
```
