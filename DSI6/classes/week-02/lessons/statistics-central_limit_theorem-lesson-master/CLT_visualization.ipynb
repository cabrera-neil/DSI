```python
import random
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline 
plt.style.use('fivethirtyeight')
```

### Basic normal distribution

```python
population = np.random.normal(loc=55,scale=3,size=10**5)
plt.hist(population,bins=100)
plt.show()
```

```python
X_mean = [np.mean(np.random.choice(population,100)) for _ in range(1000)]

plt.hist(X_mean,bins=30)
plt.show()
```

### Bimodal distribution

```python
population = np.concatenate([np.random.normal(loc=55,scale=3,size=10**5),
                             np.random.normal(loc=75,scale=7,size=10**5)])
plt.hist(population,bins=100)
plt.show()
```

```python
np.mean(population)
```

```python
X_mean = [np.mean(np.random.choice(population,100)) for _ in range(1000)]

plt.hist(X_mean,bins=30)
plt.show()
```

### Non normal - exponential distribution

```python
population = np.random.exponential(scale=4,size=10**5)
plt.hist(population,bins=100)
plt.show()
```

```python
np.mean(population)
```

```python
X_mean = [np.mean(np.random.choice(population,100)) for _ in range(1000)]

plt.hist(X_mean,bins=30)
plt.show()
```

### Small sample size

```python
population = np.random.exponential(scale=4,size=10**5)
plt.hist(population,bins=100)
plt.show()
```

```python
np.mean(population)
```

```python
SAMPLE_SIZE = 5

X_mean = [np.mean(np.random.choice(population,SAMPLE_SIZE)) for _ in range(1000)]
plt.hist(X_mean,bins=30)
plt.show()
```

```python
for sample_size in [3,5,7,10,20,30,50,100]:
    X_mean = [np.mean(np.random.choice(population,sample_size)) for _ in range(1000)]
    plt.hist(X_mean,bins=30)
    plt.title('Sample size = {}'.format(sample_size))
    plt.show()
```
