```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

%matplotlib inline 
plt.style.use('fivethirtyeight')
```

### Simmetric distribution

```python
simmetric_dist = np.random.normal(10,1,10000)
```

```python
plt.figure(figsize=(8,5))
plt.hist(simmetric_dist, bins=100)
plt.title('Simmetric distribution')
plt.xlabel('value')
plt.ylabel('freq')
plt.show()
```

#### distribution of mean-median

```python
diffs_simm = []
for _ in range(10000):
    x1 = np.random.choice(simmetric_dist,20)
    diffs_simm.append(np.mean(x1)-np.median(x1))
```

```python
plt.figure(figsize=(8,5))
plt.hist(diffs_simm,bins=100)
plt.axvline(0,c='r',lw=1)
plt.title('Mean-median - simmetric distribution')
plt.xlabel('mean-median')
plt.ylabel('freq')
plt.show()
```

# Skewed distribution

```python
skewed_dist = list(np.random.normal(11,1,5000)) + list(np.random.normal(13,2,5000))
```

```python
plt.figure(figsize=(8,5))
plt.hist(skewed_dist, bins=100)
plt.title('Skewed distribution')
plt.xlabel('value')
plt.ylabel('freq')
plt.show()
```

#### Check skewed vs. simmetric

```python
bins = np.linspace(6,20,100)
plt.figure(figsize=(8,5))
plt.hist(simmetric_dist, bins=bins, alpha=0.5,color='blue', label='simmetric')
plt.hist(skewed_dist, bins=bins, alpha=0.5,color='red', label='skewed')
plt.title('Simmetric distribution vs. Skewed distribution')
plt.xlabel('value')
plt.ylabel('freq')
plt.legend()
plt.show()
```

#### Distribution of mean-median

```python
diffs_skw = []
for _ in range(10000):
    x1 = np.random.choice(skewed_dist,20)
    diffs_skw.append(np.mean(x1)-np.median(x1))
```

```python
plt.figure(figsize=(8,5))
plt.hist(diffs_skw,bins=100)
plt.axvline(0,c='r',lw=1)
plt.title('Mean-median - skewed distribution')
plt.xlabel('mean-median')
plt.ylabel('freq')
plt.show()
```

# Check the two distributions of mean-median

```python
bins = np.linspace(-0.75,1.3,100)
plt.figure(figsize=(8,5))
plt.hist(diffs_simm, bins=bins, color='red',alpha=0.5, label='simmetric')
plt.hist(diffs_skw, bins=bins, color='blue',alpha=0.5, label='skewed')
plt.title('Distribution of mean-median: simmetric vs skewed')
plt.xlabel('mean-median')
plt.ylabel('freq')
plt.legend()
plt.show()
```
